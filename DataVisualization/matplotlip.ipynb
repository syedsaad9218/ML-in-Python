```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.random.randint(10,40, size=10)
y = np.random.randint(10,40, size=10)

plt.plot(x,y)
plt.title("Line Plot")
plt.ylabel("Y-axis")
plt.xlabel("X-axis")
plt.show()
```

```python
x = np.random.randint(10,40, size=10)
y = np.random.randint(10,40, size=10)

plt.scatter(x,y)
plt.title("Scatter Plot")
plt.ylabel("Y-axis")
plt.xlabel("X-axis")
plt.show()
```

```python
x = np.random.randint(10,40, size=10)
y = np.random.randint(10,40, size=10)

plt.bar(x,y)
plt.title("Bar Plot")
plt.ylabel("Y-axis")
plt.xlabel("X-axis")
plt.show()
```

```python
x = np.random.randint(1,50, size=30)

plt.hist(x, bins=10)
plt.title("Histogram")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
x = ['thus', 'fri', 'sat', 'sun', 'mon', 'tue', 'wed', 'thus']
y = np.random.randint(150, 250, size=8)

plt.scatter(x,y)
plt.title("Scatter Plot", fontweight='bold')
plt.ylabel("Y-axis")
plt.xlabel("X-axis")
plt.show()
```

```python
cars = ['BMW', 'Volvo', 'Ford', 'Mazda']
data = [3, 7, 5, 2]

plt.pie(data, labels=cars, autopct='%1.1f%%')
plt.title("Car Data", fontweight='bold')
plt.axis('equal')
plt.show()
```

```python
data = np.random.randint(10, 30, (3, 7))

plt.boxplot(data)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Box Plot", fontweight='bold')
plt.show()
```

```python
data = np.random.randint(10, 30, (7, 7))

plt.imshow(data, cmap='YlGn')

plt.colorbar()
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Heatmap", fontweight='bold')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

mean = 10
std_dev = 2
size = 3000

data = np.random.normal(mean, std_dev, size)

sns.histplot(data, kde=True, stat="density", bins=30, color='skyblue', linewidth=1.5)

plt.title(f"Normal distribution (mean={mean}, std_dev={std_dev})")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```
